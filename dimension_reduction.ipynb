```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
iris_df = sns.load_dataset('iris')
iris_df
```

```python
iris_df = sns.load_dataset('iris')
from sklearn.decomposition import PCA

# Fix 1: Exclude the 'species' column as PCA requires purely numeric data
pca = PCA(n_components=2)
# We drop 'species' from the dataframe before fitting
transformed = pca.fit_transform(iris_df.drop(columns='species'))

# Fix 2: Corrected the variable name from 'x_transformed' to 'transformed'
sns.scatterplot(x=transformed[:, 0], y=transformed[:, 1], hue=iris_df['species'])
plt.title('PCA of Iris Dataset')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

**t_distributed Stochastic Neighbor Embedding**

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 2, random_state = 42)
transformed = tsne.fit_transform(iris_df.drop(columns='species'))
sns.scatterplot(x = transformed[:,0], y = transformed[:,1], hue = iris_df['species'])
```
